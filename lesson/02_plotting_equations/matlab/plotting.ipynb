# Basic plotting

Matlab has a convenient built-in plotting system. It was originally the model for the popular Python plotting package, `matplotlib.pyplot`.

Create a new figure with the `figure()` function. Then use the `plot()` function to plot two equal-length vectors on a two-dimensional plane.


```matlab
x = linspace(-1,5,1000);

y = 2*x.^3 - 10*x.^2 + 5*x - 2;

figure()
plot(x,y)
```

We can plot some horizontal and vertical lines (such as the axes) by using `yline` and `xline`. `k` is shorthand for black color, and '--' denotes a dashed line.

```matlab
figure()
plot(x,y)
yline(0,color='y',linestyle='--')
xline(0,color='b',linestyle='--')
```

We can plot as many functions as we like on the same plot by using `hold on` and `hold off`. If we don't specify the color, Matlab will automatically cycle through a standard color palette:

```matlab
x = linspace(-1,5,1000);

y = 2*x.^3 - 10*x.^2 + 5*x - 2;
z = -0.5*x.^3 + 3*x.^2 - 10*x +1;
q = 5*sin(5*x);

figure()
hold on
plot(x,y)
plot(x,z)
plot(x,q)
yline(0,color='k',linestyle='--')
xline(0,color='k',linestyle='--')
hold off
```

We can use `xlim` and `ylim` to limit what the plot window shows us. In the case of functions that tend towards infinity, this can be important!

```matlab
x = linspace(-1,5,1000);

y = 2*x.^3 - 10*x.^2 + 5*x - 2;
z = -0.5*x.^3 + 3*x.^2 - 10*x +1;
q = 5*sin(5*x);
r = tan(x);

figure()
hold on
plot(x,y)
plot(x,z)
plot(x,q)
plot(x,r)
yline(0,color='k',linestyle='--')
xline(0,color='k',linestyle='--')
title('without limiting plot window')
hold off

figure()
hold on
plot(x,y)
plot(x,z)
plot(x,q)
plot(x,r)
yline(0,color='k',linestyle='--')
xline(0,color='k',linestyle='--')
xlim([min(x),max(x)])
ylim([-25,25])
title('with limiting plot window')
holf off
```

You can also label the axes, and make a legend by labeling any formulas you want included with a `'DisplayName'` and then calling the `legend()` function:

```matlab

figure()
hold on
plot(x,y,'DisplayName','f(x) = 2 x^3 - 10 x^2 + 5 x - 2')
plot(x,z,'DisplayName','f(x) = -0.5 x^3 + 3 x^2 - 10 x +1')
plot(x,q,'DisplayName','f(x) = 5 \cdot sin ( 5 x )')
plot(x,r,'DisplayName','f(x) = tan x')
yline(0,'HandleVisibility','off',color='k',linestyle='--')
xline(0,'HandleVisibility','off',color='k',linestyle='--')
xlim([min(x),max(x)])
ylim([-25,25])
legend()
title('example with 4 functions, axis labels, and a legend')
hold off
```
