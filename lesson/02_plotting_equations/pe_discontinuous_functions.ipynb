# Practical exercise: plotting discontinuous functions

For the following, write up the formula in $\LaTeX$ and then depict it on a Cartesian plane. Identify and write up any discontinuities and any limits associated with those discontinuities.

 1. The Heaviside Step Function
 2. The ratio between $x^3$ and $x^2 - 2 x + 1$.
 3. The ratio between $x^2 - 1$ and $x - 1$
 4. The tangent function.


## Heaviside Step Function

$$f(x) = \left \{ \begin{matrix*}[r l] 0 & \text{for } x < 0\\
\frac{1}{2} & \text{for } x = 0\\
1 & \text{for } x > 0\end{matrix*} \right.$$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2,2,1000)

y0 = np.ones(1000)*0
y1 = np.ones(1000)*1

y0[x >= 0] = float('nan')
y1[x <= 0] = float('nan')

plt.figure()
plt.plot(x,y0,'k')
plt.plot(x,y1,'k')
plt.plot([0],[1/2],'ko')
plt.plot([0],[0],'ko',fillstyle='none')
plt.plot([0],[1],'ko',fillstyle='none')
plt.show()
```

## Polynomial fraction #1

$$ f(x) = \frac{x^3}{x^2 - 2x + 1} $$

1 vertical asymptote/infinity discontinuity: $\lim_\limits{x \to 1} f(x) = \infty$



```python
x = np.linspace(-5,5,1000)


y = (x**3)/(x**2 - 2*x + 1)

plt.figure()
plt.plot(x,y,label='$ f(x) = \\frac{x^3}{x^2 - 2x + 1} $')
plt.ylim([-10,100])
plt.xlim([-5,5])
plt.axvline(1,color='k',linestyle='--',label='vertical asymptote')
plt.legend()
plt.show()
```

## Polynomial fraction #2

$$ f(x) = \frac{x^2-1}{x-1} $$

1 vertical asymptote/infinity discontinuity: $\lim_\limits{x \to 1} f(x) = \infty$



```python
x = np.linspace(-10,10,1000)



y = (x**2-1)/(x-1)

restriction_0 = x <= .85
restriction_1 = x >= 1.2

plt.figure()
plt.plot(x[restriction_0],
         y[restriction_0],
         'b-',
         label='$ f(x) = \\frac{x^2-1}{x-1} $')
plt.plot(x[restriction_1],
         y[restriction_1],
         'b-')
plt.ylim([-10,100])
plt.xlim([-10,10])
plt.plot([1],[2],'ko',fillstyle='none',label='point discontinuity')
plt.axvline(1,color='k',linestyle=':',alpha=.5)
plt.axhline(2,color='k',linestyle=':',alpha=.5)
plt.legend()
plt.show()
```
